```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries loaded successfully")
```

```python
df = pd.read_csv(r"C:\Users\Aleeyarh\Downloads\Superstore.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Category'].value_counts()
```

```python
df['Sales'].hist(figsize=(8,5))
plt.title('Distribution of Sales')
plt.show()
```


Most sales transactions are concentrated at lower values, while a few transactions have very high sales.

```python
sns.boxplot(x=df['Sales'])
plt.title('Boxplot of Sales')
plt.show()
```

The boxplot shows several outliers, indicating some sales transaction are much larger than the majority

```python
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title('Sales vs Profit')
plt.show()
```

There is generally a positive relationship between sales and profit,although some high sales transactions generated low profit

```python
numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True)
plt.show()
```

The heatmap shows the strenght of relationship between numerical variables. Sales and profit appears positively correlated

```python
sns.pairplot(numeric_df)
plt.show()
```

The pairplot reveals patterns and relationship among numerical variables and helps identify trends and potential outliers

 Summary of Findings

1. Most sales transactions are concentrated at lower values.
2. Several outliers exist in the Sales data.
3. Sales and Profit show a positive relationship.
4. Numerical variables exhibit varying levels of correlation.
5. The dataset contains useful patterns for business decision-making.
